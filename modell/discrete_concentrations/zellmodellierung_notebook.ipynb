Wir betrachten zwei Protonenkonzentrationen: Zum einen die innerhalb der Zelle, zum anderen außerhalb. Wir haben somit zwei "Protonenpools": Einen abgeschlossenen innerhalb der Zelle mit definiertem Volumen $V_{int}$ und physischen Grenzen, und einen im Volumen außerhalb der Zelle $V_{ext}$. Für die äußere Protonenkonzentration nehmen wir an, dass sich die Teilchen über ein Effekivvolumen ausbreiten (etwa dem Volumen, das der Zelle innerhalb der Nährlösung zu Verfügung steht). Wir ordnen also den Volumina feste Werte zu, mit denen wir die Konzentrationen berechnen: Die Protonenkonzentrationen $c_{int}(t)$ innerhalb und $c_{ext}(t)$ außerhalb der Zelle. 

Wir vereinfachen das System weiter, indem wir festlegen, dass sich die Protonen gleichmäßig über die jeweiligen Volumina verteilen. In Realität wäre dies nicht der Fall. Es gäbe beispielsweise elektrostatische Kräfte, die besonders an der Zellmembran eine Rolle spielen. Zur genauen Betrachtung des System muss räumlich unterschieden werden und es ist nötig, weitere Ionenbewegungen, etwa Chlorid-Ionen, zu betrachten. In unserem Modell interessieren wir uns nur für Änderungen der Reaktionsgeschwindigkeit durch eine Änderung der Konzentrationen der beteiligten Stoffe und keine weiteren Kräfte oder Einflussfaktoren. Wir können die Konzentrationsänderung innerhalb bzw. außerhalb der Zelle also berechnen durch:

$$\frac{\partial c_{int}}{\partial t} = \frac{v}{V_{int}} \qquad \mathrm{und} \qquad \frac{\partial c_{ext}}{\partial t} = - \frac{v}{V_{ext}}$$

Wir wollen also wissen, wie viele Protonen pro Sekunde (angegeben mit einer "Protonengeschwindigkeit" $v$) bei einem bestimmten Gefälle $\delta c = c_{int} - c_{ext}$ über die Membran wandern. Dabei gibt es drei Hauptströme:

Den Protonenfluss über die Atmungskette $v_{Atm}$, <br />
den Protonenfluss über die ATP-Synthase $v_{ATP}$, <br />
und den Protonenfluss über unseren Kanal $v_{Kanal}$. <br />
Hinzu kommt noch eine Basisrate über den Rest der Membran $v_{Membran}$.

Die Ströme lassen sich aufaddieren zur einem Gesamtstrom $v$:

$$\implies v \:= \sum_{i\, \in \, K} v_i \: = \: v_{Atm} + v_{ATP} + v_{Kanal} + v_{Membran} $$

wobei wir die Menge der Membrankomponenten $K\: = \:(Atm,\:ATP,\:Kanal,\:Membran)$ nennen und die Raten $v_i$ nach Innen gerichtet sind.

#### Herleitung einer Formel zur Beschreibung des Flusses durch den Kanal
Die Gesamtreaktion die abläuft, ist: $H^+_{in} + E_1 \leftrightarrow C_1 \leftrightarrow C_2 \leftrightarrow E_2 + H^+_{out}$ mit den Reaktionsfaktoren $k_1$ u. $k_{-1}$ für die erste, $k_2$ u. $k_{-2}$ für die zweite sowie $k_3$ u. $k_{-3}$ für die dritte Teilreaktion.

Man kann nun die Änderungsraten der einzelnen Teiledukte und Produkte beschreiben: 
$$ \frac{dH^+_{in}}{dt} = -k_1 H^+_{in} E_1 + k_{-1} C_1$$
$$ \frac{dE_1}{dt} = -k_1 H^+_{in} E_1 + k_{-1} C_1$$
$$ \frac{dC_1}{dt} = k_1 H^+_{in} E_1 + k_{-2} C_2 - k_2 C_1$$
$$ \frac{dC_2}{dt} = k_{-3} E_2 H^+_{out} + k_{2} C_1 - k_{-2} C_2 - k_3 C_2$$
$$ \frac{dE_2}{dt} = k_3 C_2 - k_{-3} E_2 H^+_{out}$$
$$ \frac{dH^+_{out}}{dt} = k_3 C_2 - k_{-3} E_2 H^+_{out}$$

Wir wissen, dass die Gesamtmenge an Enzym konstant bleibt: $E_1 + C_1 = E^0_1 + C^0_1$, sowie $E_2 + C_2 = E^0_2 + C^0_2$ Somit können wir in den obrigen Gleichungen $E$ ersetzen: $E_1 = E^0_1 + C^0_1 - C_1$ beziehungsweise $E_2 = E^0_2 + C^0_2 - C_2$
$$ \frac{dC_1}{dt} = k_1 H^+_{in} (E^0_1 + C^0_1 - C_1) + k_{-2} C_2 - k_2 C_1$$
$$ \frac{dC_2}{dt} = k_{-3} (E^0_2 + C^0_2 - C_2) H^+_{out} + k_{2} C_1 - k_{-2} C_2 - k_3 C_2$$

Die obrigen Differentialgleichungen können wir lösen indem mir annehmen, dass sich auf der von uns betrachteten Zeitskala sowohl die Protonenkonzentration $H^+_{out}$ und $H^+_{in}$ als auch die Konzentration an gebundenen Kanal-Protonen-Komplexen nicht ändert. Wir suchen also $C_1$ und $C_2$ sodass gilt: $\frac{dC_1}{dt} = 0$ und $\frac{dC_2}{dt} =0$
$$ 0 = k_1 H^+_{in} (E^0_1 + C^0_1 - C_1) + k_{-2} C_2 - k_2 C_1$$
$$ \Rightarrow C_1 = \frac{k_1 H^+_{in}(E^0_1 + C^0_1) + k_{-2} C_2}{k_1 H^+_{in} + k_2}$$

$$ 0 = k_{-3} E_2 H^+_{out} + k_{2} C_1 - k_{-2} C_2 - k_3 C_2$$
$$ \Rightarrow C_2 = \frac{k_{-3} H^+_{out} (E^0_2 + C^0_2)}{k_{-2} - k_2C_1 - k_3 + k_{-3}H^+_{out}}$$

* $ v = k_2 C_1 - k_{-2} C_2$

Anmerkung => Erstmal einfachstes Modell implementieren, sollte schnell gehen.

## Implementierung

Für die Implementierung führen wir hier die "Ortskoordinate" $x$ ein, welche die diskreten Werte $0$ oder $1$ beziehungsweise $innen$ und $außen$ annehmen. Somit können Konzentration $c$, Volumen $V$, ... im Code als $c[0]$, $V[0]$, ... verwendet werden.

```python
import numpy as np
from scipy.integrate import odeint
```

## Parameter

### Anfangsbedingungen und Auflösung

```python
# Zeitraum und Integrationsschritte
# Wenn man die Schritte größer als die Zeit macht, dann bekommt man Probleme mit der Rate.
# Diese wird über die Anzahl der Schritte differenzert und nicht über die Zeit, also kommen kleinere Werte raus
steps = 1000
time = 1
```

```python
# Innen- und Außenvolumen
V = [0.01,0.01] #Initialisiere V mit beliebigen Anfangswerten (zum testen verwendet)
r_cell = 1.5 * 10**(-3)
V[0] = (4/3) * np.pi * (r_cell**3) #Berechne Volumen der Zelle näherungsweise als Kugel mit Radius r_cell
V[1] = 1 #Außenvolumen, das sehr viel größer als das Zellvolumen ist
```

```python
# Anfangskonzentrationen
#pH_0 = [7, 0] #Setze pH innen auf 7, außen auf 0.
#c_0 = [np.e**(-pH_0[0]), np.e**(-pH_0[1])] #Rechne pH in H+-Konzentration um.
c0_in = np.e**(-7)
c0_out = np.e**(-5.5)
# Testing
# Array c hat viel mehr als nur die Konzentration
# c[0] = c innen
# c[1] = c außen
# c[2] = v atp-synthase
# c[3] = v atmungskette
# c[4] = v kanal
c_0 = [c0_in,c0_out,0,0.01,0.01]
```

## Funktionen

### Protonenfluss über die ATP-Synthase

Fluss über die ATP-Synthase, wie beschrieben von Nath et al.

```python
def v_ATP(c, vmax=2, k1=2, k2=2, N=1):
    # Das Ganze ist vertauscht
    if c[0] < c[1]:
        cint = c[0]
        cext = c[1]
        direction = 1.0
    else:
        cint = c[0]
        cext = c[1]
        direction = 0
    
    kskrE0 = 57 
    k1k2 = 900
    k1kre0kt = 10**(-5.18)
    v = kskrE0*(cext/cint)/((cext/cint) + k1k2 + k1kre0kt/cint)
    return v * direction
    
    #v = vmax * c[0] / (c[0] + k1 * (1 + c[1] / k1))
    #return v
```

### Protonenfluss über den Kanal

```python
def v_Kanal(c):
    k1a = 1
    k1b = 1 
    k2a = 1
    k2b = 1
    k3a = 1
    k3b = 1
    kanalkonz = 1
    cin = c[0]
    cout = c[1]
    C2 = ( k3b*kanalkonz - (k2a*k1a*kanalkonz*cin)/(k1a*cin + k2a*k1b) )/( k3b + k2a + k3a - ((k1a*k2b)/(k1a*cin + k1a + k1b)))
    v =  C2 * (k3a + k3b*cout) - k3b*kanalkonz
    return -v
```

### Protonenfluss über die Atmungskette

```python
def v_Atm(c, vmax=1):
    #v = 0.5
    v = -vmax * c[0] / (10**(-2) + c[0])
    #Einfach MM-Kinetik um schnell ansteigendes und dann schnell konstant bleibendes Verhalten zu bekommen
    return v
```

### Zusammenführung zum Gesamtfluss und Verrechnung zur Konzentration

```python
def cdot(c, t):
    vATP = v_ATP(c)
    vAtm = v_Atm(c, vmax = 3.5)
    vKanal = v_Kanal(c)
    v = vATP + vAtm + vKanal
    dcdt = [v/V[0], -v/V[1], vATP, vAtm, vKanal]
    return dcdt
```

# Räumliche Animation

```python
# Zum Animieren in diesem notebook
%matplotlib notebook
from matplotlib import animation
from matplotlib.animation import FuncAnimation

# Array für Bild definieren
height = 400
width = 400

# half-height
hh = int(height/2)
# maximum concentration
maxconc = max([max([solution[x][0] for x in range(0,len(solution))]), 
               max([solution[x][1] for x in range(0,len(solution))])])

# Array mit Wert 0 initialisieren
image = np.zeros((height, width))
frame = image
```

```python
%matplotlib notebook

# Figure und Animation
memanim = plt.figure()

# Membran einzeichnen
quer, mem1, mem2 = [0, width], [-5, -5], [+5, +5]
plt.plot(quer, mem1, quer, mem2, color='green', lw=3, marker = None)

im=plt.imshow(image, interpolation='none', cmap='Blues', vmin=0, vmax=2,
              aspect='auto', extent=[0,width,-hh,hh])
plt.xticks([])

    
# Initialisierungsfunktion: Plottet für jeden Frame den Hintergrund
def init():
    return [image]

# Animationsfunktion: Wird sequenziell vom Animator aufgerufen
def animate(i):
    frame[0 : hh-5, 0 : width] = solution[i][1]/maxconc*1.5+0.5
    frame[hh+5 : height, 0 : width] = solution[i][0]/maxconc*1.5+0.5
    
    im.set_array(frame)
    return [im]

# Animator aufrufen
# blit=True damit nur veränderte Pixel neu gesetzt werden
anima = animation.FuncAnimation(memanim, animate, init_func=init, frames=steps, 
                               interval=int(time/steps*1000), blit=True)

# animation als *.mp4 speichern
#anima.save('./figures/animation_mem_conc.mp4', fps=30, 
#           extra_args=['-vcodec', 'libx264'])

# animation als *.gif speichern
# Unter Windows muss dafür evtl. ImageMagick installiert
# und der convert_path definiert werden, um *.gif s zu speichern
#plt.rcParams['animation.convert_path'] = '<path-to>/magick.exe'

writer = animation.ImageMagickFileWriter()
writer.fps = 30
#anima.save('./figures/animation_mem_conc.gif', writer=writer)
```
